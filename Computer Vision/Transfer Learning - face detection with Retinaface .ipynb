face detection model RetinaFace from InsightFace.

A modern face recognition pipeline consists of 4 common stages: detect, align, represent and verify. Detection and alignment are early and very important stages. RetinaFace handles those early stages with a high confidence score. 

# Structure

RetinaFace is mainly based on an academic study: RetinaFace: Single-stage Dense Face Localisation in the Wild. Model design is based on feature pyramids. Context modules come after the independent context modules. 

If priority is confidence, then consider to use mtcnn, retinaface or dlib; on the other hand, if priority is speed, consider to use opencv or ssd.

# Face Detection

RetinaFace offers a face detection function. It expects an exact path of an image as input.

Then, it will return the facial area coordinates and some landmarks (eyes, nose and mouth) with a confidence score.

The default threshold value is set to 0.9. You can decrease it if you want to detect faces in low resolution. 

```python
result = image_processing.face_detection_retinaface(
    image_data=r'E:\gitlab\dataset\images\t1.jpg'
    ,threshold=0.9,nms_threshold = 0.4,decay4=0.5)
```
